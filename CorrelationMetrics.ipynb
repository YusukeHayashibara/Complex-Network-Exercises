## Q2. Calcule a correlação de Pearson entre a medida betweeness centrality e grau para a rede hamsterster. Use:



G = G.to_undirected()

G.remove_edges_from(nx.selfloop_edges(G))

Gcc = sorted(nx.connected_components(G), key=len, reverse=True)

G = G.subgraph(Gcc[0])

G = nx.convert_node_labels_to_integers(G, first_label=0)

```python
G = nx.read_edgelist('/Users/yusukehayashibara/Downloads/data2Q/hamsterster.txt')
plt.figure(figsize=(12,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, node_color="lightgray", node_size=100, with_labels=False)
```

```python
G = G.to_undirected()

G.remove_edges_from(nx.selfloop_edges(G))

Gcc = sorted(nx.connected_components(G), key=len, reverse=True)

G = G.subgraph(Gcc[0])

G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
degrees = dict(G.degree())
graus = np.array(list(degrees.values()))
```

```python
bc = nx.betweenness_centrality(G)
betweeness = np.array(list(bc.values()))
```

```python
rho = np.corrcoef(graus, betweeness)[0, 1]
print(f'Pearson correlation coefficient:): {rho:.3f}')
```

```python
G = nx.read_edgelist('/Users/yusukehayashibara/Downloads/data2Q/USairport500.txt')
plt.figure(figsize=(12,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, node_color="lightgray", node_size=100, with_labels=False)
```

```python
G = G.to_undirected()

G.remove_edges_from(nx.selfloop_edges(G))

Gcc = sorted(nx.connected_components(G), key=len, reverse=True)

G = G.subgraph(Gcc[0])

G = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
#k-core
KC= dict(nx.core_number(G))
print('k-core:', KC)
KC = list(KC.values())
```

```python
#Closeness centrality
CLC = dict(nx.closeness_centrality(G))
CLC = list(CLC.values())
```

```python
from scipy.stats import spearmanr
rho_spearman, p_value = spearmanr(KC, CLC)

print(f'Spearman correlation coefficient:: {rho_spearman:.3f} ')
```
